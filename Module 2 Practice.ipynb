```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("/content/companies.csv")
```

```python
df.head()
```

```python
state_counts = df['state'].value_counts().reset_index()
state_counts.columns = ['state', 'count']
plt.figure(figsize=(18, 8))
plt.bar(x=state_counts["state"], height=state_counts["count"])
plt.title("Number of Companies in Each State")
plt.xlabel("State")
plt.ylabel("Count")
```

```python
opening_counts = df.groupby('Industry')['job_openings'].sum().reset_index()
opening_counts.columns = ['Industry', 'Job_Openings']
plt.figure(figsize=(20, 8))
plt.bar(x=opening_counts ["Industry"], height=opening_counts["Job_Openings"])
plt.title("Sum of Job Openings by Industry")
plt.xlabel("State")
plt.ylabel("Count")
```

```python
employees_revenue = df.groupby('estimated_revenues')['employees'].sum().reset_index()
employees_revenue.columns = ['estimated_revenues', 'employees']
plt.figure(figsize=(20, 8))
plt.scatter(employees_revenue ["estimated_revenues"], employees_revenue["employees"])
plt.title("Employees to Estmated Revenues")
plt.xlabel("estimated_revenues")
plt.ylabel("employees")
plt.grid(True)
plt.show()
```

```python
companies_founded = df.groupby('founded').size().reset_index(name='counts')
companies_founded.columns = ['founded', 'counts']
plt.figure(figsize=(20, 8))
plt.hist(companies_founded["founded"])
plt.title("Number of Companies Founded each Year")
plt.xlabel("Founded Year")
plt.ylabel("Number of Companies")
plt.grid(True)
plt.show()
```
